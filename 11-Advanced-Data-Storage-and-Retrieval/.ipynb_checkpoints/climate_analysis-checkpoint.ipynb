# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect,distinct
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
#Findout all the columns for the given table
inspector = inspect(engine)
columns = inspector.get_columns('station')
columns
```

```python
#Findingout the most recent record
results = session.query(Measurement.id,Measurement.station,Measurement.date,Measurement.prcp,Measurement.tobs).order_by(Measurement.date.desc()).first()
results
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Calculate the date 1 year ago from today
latest_record = session.query(Measurement.id,Measurement.station,Measurement.date,Measurement.prcp,Measurement.tobs).order_by(Measurement.date.desc()).first()
latest_record_date = dt.datetime.strptime(latest_record.date, "%Y-%m-%d").date()

dateYearAgo=latest_record_date - dt.timedelta(days=(365)) #Getting the date 12months before the most recent record
dateYearAgo
# Perform a query to retrieve the data and precipitation scores
results = session.query(Measurement.date,Measurement.prcp).filter(Measurement.date>dateYearAgo).all()
results
# Save the query results as a Pandas DataFrame and set the index to the date column
prcp_df = pd.DataFrame(results)
prcp_df['date'] = pd.to_datetime(prcp_df['date'])
prcp_df.set_index('date',inplace=True)
prcp_df.head()
# Sort the dataframe by date
prcp_df = prcp_df.sort_values(by=['date'])
prcp_df.head()
# Use Pandas Plotting with Matplotlib to plot the data
plt.bar(prcp_df.index,prcp_df['prcp'],label='precipation')
plt.xlabel('date')
# Rotate the xticks for the dates
plt.xticks(rotation=45)
plt.legend(loc='upper center')
```

```python
# Write a function called `calc_temps` that will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
trip_temp_summary=calc_temps('2017-01-01','2017-02-01')

temperature_summary=list(np.ravel(trip_temp_summary)) #convert sql tuple result set to a normal list

temperature_summary
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
min_temp = temperature_summary[0]
max_temp = temperature_summary[2]
avg_temp = temperature_summary[1]
yerror = max_temp - min_temp


```

```python
fig,ax = plt.subplots(figsize=(5,10))
ax.bar('', avg_temp, yerr=yerror , color='sandybrown' )
ax.set_xticks([1]) 
plt.ylabel("Temp (F)")
plt.title("Trip Avg temperature")
plt.tight_layout()
  
plt.show()
```

```python
# Calculate the rainfall per weather station for your trip dates using the previous year's matching dates.
# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation
def calculate_rainfaill_per_station(start_date,end_date):
    
    prev_year_start_date = dt.datetime.strptime(start_date, "%Y-%m-%d").date()-dt.timedelta(days=365)
    
    prev_year_end_date = dt.datetime.strptime(end_date, "%Y-%m-%d").date()-dt.timedelta(days=365)
    
    sel = [Station.station, 
           Station.name, 
           Station.latitude, 
           Station.longitude, 
           Station.elevation,
           func.avg(Measurement.prcp).label('avg_prcp')]
    
    rainfall_results = session.query(*sel).filter(Measurement.station == Station.station).\
                        filter(Measurement.date>=prev_year_start_date).\
                        filter(Measurement.date<=prev_year_end_date).\
                        group_by(Station.station).\
                        order_by(func.avg(Measurement.prcp)).all()
    return rainfall_results
                                                                                                     
    
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
# calculate the daily normals for your trip
# push each tuple of calculations into a list called `normals`
# Set the start and end date of the trip
# Use the start and end date to create a range of dates
# Stip off the year and save a list of %m-%d strings
# Loop through the list of %m-%d strings and calculate the normals for each date
def get_trip_days_list(trip_start_date,trip_end_date):
    converted_end_date = dt.datetime.strptime(trip_end_date, "%Y-%m-%d").date()
    convert_start_date = dt.datetime.strptime(trip_start_date, "%Y-%m-%d").date()
    trip_duration=converted_end_date-convert_start_date
    
    days_list = [(converted_end_date-dt.timedelta(days=x))  for x in range(0,trip_duration.days+1)]
    return days_list
    
```

```python
def get_daily_normals_list(trip_start_date,trip_end_date):
    
    tuple_list=[]
    daystring_list =[d.strftime("%m-%d") for d in get_trip_days_list(trip_start_date,trip_end_date)]
    for day in daystring_list:
        tuple_list.append(daily_normals(day)[0])
    return tuple_list
        

daily_normals_list= get_daily_normals_list('2017-01-01','2017-02-01')
daily_normals_list
```

```python
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index
daily_normals_df = pd.DataFrame(daily_normals_list,columns=['tmin','tavg','tmax'])
daily_normals_df['date'] =get_trip_days_list('2017-01-01','2017-02-01')

daily_normals_df=daily_normals_df.sort_values('date')
daily_normals_df.set_index('date',inplace=True)
daily_normals_df
```
